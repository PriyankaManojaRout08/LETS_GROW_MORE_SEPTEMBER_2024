# Subplots which includes Sactterplot,Swarmplot,Lineplot and Barplot

```python
plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
sns.scatterplot(x='sepal_length',y='species',color='green',data=df)

plt.subplot(2,2,2)
sns.swarmplot(x='sepal_length',y='species',color='yellow',data=df)

plt.subplot(2,2,3)
sns.lineplot(x='sepal_length',y='species',color='purple',data=df)

plt.subplot(2,2,4)
sns.barplot(x='sepal_length',y='species',color='pink',data=df)
```

# Pair Plot

```python
sns.pairplot(df,hue='species')
plt.show()
```

```python
sns.pairplot(df,hue='sepal_length')
plt.show()
```

# Heatmap

```python
sns.heatmap(df.corr())
```

# Violin Plot

```python
sns.violinplot(x ="sepal_width",y ="species",data =df)
```

# Joint Plot

```python
sns.jointplot(x = "sepal_length", y = "species", data = df)
```

# Stripplot

```python
sns.stripplot(x = "sepal_length", y = "species", data = df)
```
