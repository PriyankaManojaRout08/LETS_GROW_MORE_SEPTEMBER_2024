TASK 4-Image to Pencil sketch with Python

# Importing Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

# Reading the image

```python
image = cv2.imread("/content/flowers.jpg")
from google.colab.patches import cv2_imshow
#title("Original Image")
cv2_imshow(image)
cv2.waitKey(0)
```

# Converting the Original image to Greyscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

# Inverting the new grayscale image

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

# Gaussian Function in OpenCV to blur image

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

# Converting the image into a Pencil sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

# Original image and the Pencil sketch

```python
cv2_imshow(image)
cv2_imshow( pencil_sketch)
cv2.waitKey(0)
```
